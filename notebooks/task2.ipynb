```python
X = db_scaled.values
from sklearn.cluster import KMeans
#fitting kmeans to dataset
kmeans = KMeans(n_clusters=3, init= 'k-means++', n_init = 10, max_iter = 300, random_state= 0 )
Y_kmeans = kmeans.fit_predict(X)

cluster_1 = pd.DataFrame(X[Y_kmeans == 0], columns=cluster_columns)
cluster_2 = pd.DataFrame(X[Y_kmeans == 1], columns=cluster_columns)
cluster_3 = pd.DataFrame(X[Y_kmeans == 2], columns=cluster_columns)
```

```python
def top_user_per_app(df:pd.DataFrame, Id:str,col:str,count=10):
    return df.groupby(Id)[col].sum().sort_values(ascending=False).reset_index()[:count]
```

Plot the top 3 most used applications using appropriate charts:

```python
apps = db.columns.to_list()
del apps[0:3]
del apps[-3:]
```

```python
total=[]
for col in apps:
    total.append(db[col].sum())
    
```

```python
app_total = pd.DataFrame(data=[apps,total])
app_total=app_total.transpose()
app_total.rename(columns={0:'app',1:'total data'}, inplace=True)
```

```python
app_total.sort_values(by='total data', ascending=False,inplace=True)
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(app_total['app'], app_total['total data'])
plt.title('Top Applications' , size=20)
plt.xticks(rotation=75, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Applications', fontsize=16)
plt.ylabel('Total Data', fontsize=16)
plt.show()
```
