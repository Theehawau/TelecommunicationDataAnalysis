Normalizing:

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
```

```python
column_names_to_normalize = ['Dur. (ms)','Number of sessions','Total (Bytes)']
X = db[column_names_to_normalize].values
x_scaled = normalizer.fit_transform(X, y =None)
db_temp = pd.DataFrame(x_scaled, columns=column_names_to_normalize, index = db.index)
db[column_names_to_normalize] = db_temp
```

```python
 db_temp.head()
```

```python
sns.set(rc={'figure.figsize':(11,8)})
sns.distplot(db['Dur. (ms)'], kde=True, color='blue')
plt.title(f'Distribution of Normalized Duration', size=20, fontweight='bold')
plt.show()
```

K-means Clustering:

```python
#determining number of clusters with elbow method
X = db_temp.values
from sklearn.cluster import KMeans
#fitting kmeans to dataset
kmeans = KMeans(n_clusters=3, init= 'k-means++', n_init = 10, max_iter = 300, random_state= 0 )
Y_kmeans = kmeans.fit_predict(X)

cluster_1 = pd.DataFrame(X[Y_kmeans == 0], columns=column_names_to_normalize)
cluster_2 = pd.DataFrame(X[Y_kmeans == 1], columns=column_names_to_normalize)
cluster_3 = pd.DataFrame(X[Y_kmeans == 2], columns=column_names_to_normalize)
```

Aggregate user total traffic per application and derive the top 10 most engaged users per application


```python
def top_user_per_app(df:pd.DataFrame, Id:str,col:str,count=10):
    return df.groupby(Id)[col].sum().sort_values(ascending=False).reset_index()[:count]
```

```python
top_user_per_app(db,'MSISDN/Number','HTTP (Bytes)')
```

```python
top_user_per_app(db,'MSISDN/Number','Social Media (Bytes)')
```

```python
top_user_per_app(db,'MSISDN/Number','Netflix (Bytes)')
```
